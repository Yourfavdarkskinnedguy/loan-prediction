```python
encoder= LabelEncoder()
scaler= StandardScaler()
```

```python
data['person_gender']=encoder.fit_transform(data['person_gender'])
data['person_education']=encoder.fit_transform(data['person_education'])
data['person_home_ownership']=encoder.fit_transform(data['person_home_ownership'])
data['loan_intent']=encoder.fit_transform(data['loan_intent'])
data['previous_loan_defaults_on_file']=encoder.fit_transform(data['previous_loan_defaults_on_file'])
```

```python
data.info()
```

```python
data.head()
```

```python
from sklearn.utils import resample
```

```python
 minority_class= data[data['loan_status']==1]
 majority_class= data[data['loan_status']==0]
```

```python
majority_class
```

```python
majority_class_resampled= resample(
    majority_class,
    n_samples=len(minority_class),
    random_state=42
)
```

```python
new_data= pd.concat([minority_class, majority_class_resampled]).sample(frac=1, random_state=2).reset_index(drop=True)
```

```python
new_data.head(10)
```

```python
X= new_data.drop(['loan_status'], axis=1)
```

```python
Y= new_data['loan_status']
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=6)
```

```python
X_train_scaled= scaler.fit_transform(X_train)
X_train= pd.DataFrame(X_train_scaled, columns=X_train.columns)
```

```python
X_test_scaled= scaler.fit_transform(X_test)
X_test= pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
model= RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy score on train data
ypred_train= model.predict(X_train)
train_acc= accuracy_score(ypred_train, Y_train)
print(train_acc)
```

```python
#accuracy score on test data
ypred_test= model.predict(X_test)
test_acc= accuracy_score(ypred_test, Y_test)
print(test_acc)
```
